```python
max_value=10000
class KNN_created:
    def __init__(self, k):
        self.k = k
        self.training_data = []
        self.training_labels = []
    
    def fit(self, training_data, training_labels):
        self.training_data = training_data
        self.training_labels = training_labels
        self.training_labels_list = np.unique(self.training_labels)

    
    def predict(self, test_data):
        
        predictions = []
        #times=0
        
        for row in test_data:
            #times+=1
            #print("=====================================")
            #print("第 ",times," 次筆" )
            
            label = self.closest(row)
            #print("預測標籤為：",label)
            predictions.append(label)
        return predictions
    
    def closest(self, new_data):#若無相同的距離則可順利回傳最終的標籤
        
        
        #print("new_data:",new_data)
        distance=[]#1/距離
        for i in range(len(self.training_data)):
            #print("第 ",i," 筆資料為：",self.training_data[i])
            return_distance=self.calculate_distance(new_data, self.training_data[i])
            #print("第 ",i," 筆資料與新資料的距離為：",return_distance)
            if return_distance:
                distance.append(1/return_distance)
            else:
                #continue
                distance.append(max_value)
        
        k_distance=[]
        k=self.k
        while k>0:
            
            for i in range(len(distance)):
                if distance[i]==max(distance):
                    k_distance.append([i,max(distance)])
                    distance[i]=0
                    k-=1
            
        #print("k_distance:",k_distance)
        #print("=====================================")
        
        label_weight=dict()
        #若label不為二元，則需修改
        label_weight[self.training_labels_list[0]]=0
        label_weight[self.training_labels_list[1]]=0
        
        #將各種類的權重加總
        for i in range(len(k_distance)):
            #print("第 ",i," 筆資料的標籤為：",self.training_labels[k_distance[i][0]])
            
            if self.training_labels[k_distance[i][0]]==self.training_labels_list[0]:
                label_weight[self.training_labels_list[0]]+=k_distance[i][1]
            if self.training_labels[k_distance[i][0]]==self.training_labels_list[1]:
                label_weight[self.training_labels_list[1]]+=k_distance[i][1]
                
        if label_weight[self.training_labels_list[0]]>label_weight[self.training_labels_list[1]]:
            return self.training_labels_list[0]
        else:
            return self.training_labels_list[1]
    
    def calculate_distance(self, a, b):#計算a,b之間的距離
        if type(a)!=type(b):
            print("a:",a,"type:",type(a))
            print("b:",b,"type:",type(b))
            return "calculate_distance func error, type(a)!=type(b)"
        return np.sqrt(np.sum(np.square(a - b)))
    
    def score(self, test_pred, test_labels):#計算準確率
        predictions = test_pred
        correct = 0
        for i in range(len(test_labels)):
            if predictions[i] == test_labels[i]:
                correct += 1
        return correct / len(test_labels)
```
