```python
# 欄位設定
x_col = ['name', 'rate', 'subject', 'income_per_month', 'buy_tablet_or_not']

x_num_col = ['rate','income_per_month']
x_cat_col = ['subject']
```

```python
df_cat = df[x_cat_col]
df_num = df[x_num_col]
```

處理數值型

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import make_pipeline
num_pl=make_pipeline(MinMaxScaler())
numcol_input=num_pl.fit_transform(df_num)
numcol_input
```

```python
pd.DataFrame(numcol_input.round(3),columns=x_num_col)
```

處理類別型

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()
cat_pl=make_pipeline(OneHotEncoder(sparse=False))
catcol_input=cat_pl.fit_transform(df_cat)
catcol_input
```

```python
cat_colname=cat_pl.named_steps['onehotencoder'].get_feature_names_out(x_cat_col)
cat_colname=list(cat_colname)
```

```python
pd.DataFrame(catcol_input.round(3),columns=cat_colname)
```

```python
num_input=np.array(numcol_input)
cat_input=np.array(catcol_input)

#name=np.array(df['name'])
input_data=np.hstack((num_input,cat_input))

pd.DataFrame(input_data,columns=x_num_col+cat_colname)
```

```python
train_x = input_data[:14]
train_y = df['buy_tablet_or_not'][0:14]
    
test_x = input_data[14:]
test_y = df['buy_tablet_or_not'][14:]
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
tree.fit(train_x,train_y)
```

```python
tree.predict(test_x)
```

```python
tree.score(test_x,test_y).round(3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y,tree.predict(test_x))
```

```python
output=tree.predict(test_x)
```

```python
wili_buy=[]
for i in range(len(output)):
    if output[i]=='y':
        wili_buy.append(df['name'][i+14])
wili_buy
```
